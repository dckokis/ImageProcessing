# Task 1
Написать программу для определения
количества точек в изображении с яркостью выше
заданного значения (T). Исходное изображение
цветное и загружается с диска

```python
img = cv.imread('img.png')
print(img)
```

```python
from PIL import Image
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
display(Image.fromarray(img))
```

```python
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
display(Image.fromarray(img_gray))
```

```python
T = int(input("Input value: "))
amount = 0
for x in range(img_gray.shape[0]):
    for y in range(img_gray.shape[1]):
        if img_gray[x, y] > T:
            amount +=1
N = amount / img_gray.size * 100
print(str(N) + " %")
```
