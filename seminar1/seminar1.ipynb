# Task 1
Написать программу для определения
количества точек в изображении с яркостью выше
заданного значения (T). Исходное изображение
цветное и загружается с диска

```python
img = cv.imread('img.png')
print(img)
```

```python
from PIL import Image

img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
display(Image.fromarray(img))
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
display(Image.fromarray(img_gray))
```

```python
T = int(input("Input value: "))
amount = 0
for x in range(img_gray.shape[0]):
	for y in range(img_gray.shape[1]):
		if img_gray[x, y] > T:
			amount += 1
N = amount / img_gray.size * 100
print(str(N) + " %")
```

# Task 2
Реализовать «вручную» цветовое выделение
значимых частей на изображении: глаз, ушей, носа кота
– с помощью примитивов отрисовки (полигонов, кругов,
эллипсов, линий).

```python
cat = cv.imread('cat.png')
cat = cv.cvtColor(cat, cv.COLOR_BGR2RGB)
display(Image.fromarray(cat))
cat.shape
```

```python
# first eye
firstEyeCoordinates = (670, 480)
# second eye
secondEyeCoordinates = (990, 483)
axesLength = (60, 60)
angle = 0
startAngle = 0
endAngle = 360

# first ear
p1 = (500, 10)
p2 = (480, 350)
p3 = (670, 200)

# second ear
p4 = (1180, 30)
p5 = (990, 220)
p6 = (1190, 440)

# nose
n1 = (800 ,640)
n2 = (880, 640)
n3 = (830, 720)

cat_with_parts = cv.ellipse(cat, firstEyeCoordinates, axesLength, angle, startAngle, endAngle, (255, 0, 0), 5)
cv.ellipse(cat_with_parts, secondEyeCoordinates, axesLength, angle, startAngle, endAngle, (255, 0, 0), 5)
cv.line(cat_with_parts, p1, p2, (0, 255, 0), 5)
cv.line(cat_with_parts, p2, p3, (0, 255, 0), 5)
cv.line(cat_with_parts, p1, p3, (0, 255, 0), 5)
cv.line(cat_with_parts, p4, p5, (0, 255, 0), 5)
cv.line(cat_with_parts, p5, p6, (0, 255, 0), 5)
cv.line(cat_with_parts, p4, p6, (0, 255, 0), 5)
cv.line(cat_with_parts, n1, n2, (0, 255, 255), 5)
cv.line(cat_with_parts, n2, n3, (0, 255, 255), 5)
cv.line(cat_with_parts, n1, n3, (0, 255, 255), 5)
```

```python
cat_with_parts = cv.cvtColor(cat_with_parts, cv.COLOR_RGB2BGR)
cv.imwrite('C:\ImageProcessing\seminar1\cat_with_parts.png', cat_with_parts)
```
