# Canny edge detection

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
cards = cv.imread('cards.png')
edges = cv.Canny(cards, 100, 200)
```

```python
plt.rcParams['figure.dpi'] = 200
plt.axis('off')
plt.imshow(cv.cvtColor(cards, cv.COLOR_BGR2RGB), cmap='gray')
plt.show()
plt.imshow(edges, cmap='gray')
plt.show()
```

```python
donut = cv.imread('donut.png')
donut_gray = cv.cvtColor(donut, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(donut_gray, 240, 255, cv.THRESH_BINARY)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
cv.drawContours(donut, contours, -1, (0, 255, 0), 3)
plt.imshow(cv.cvtColor(donut, cv.COLOR_BGR2RGB), cmap='gray')
plt.axis('off')
plt.show()
```

# Try with Canny

```python
cards = cv.imread('cards.png')
gray_cards = cv.cvtColor(cards, cv.COLOR_BGR2GRAY)
edges = cv.Canny(gray_cards, 100, 200)
ret, thresh = cv.threshold(edges, 140, 255, cv.THRESH_BINARY)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)

cv.drawContours(cards, contours, -1, (0, 255, 0), 3)
plt.imshow(cards, cmap='gray')
plt.axis('off')
plt.show()
```
