```python
import cv2 as cv
```

```python
import numpy as np
from matplotlib import pyplot as plt

sonnet = cv.imread('sonnet.png')
sonnet= cv.cvtColor(sonnet, cv.COLOR_BGR2GRAY)
processed = np.array(cv.adaptiveThreshold(sonnet, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 3))

plt.rcParams['figure.dpi'] = 200
plt.imshow(processed, cmap='gray')
plt.axis('off')
```

```python
cv.imwrite('C:/ImageProcessing/seminar2/processed.png', processed)
```
