```python
import cv2 as cv
from PIL import Image
```

```python
import numpy as np
import matplotlib.pyplot as plt

img = np.array(cv.cvtColor(np.array(cv.imread('Lena.png')), cv.COLOR_BGR2GRAY))

plt.rcParams['figure.dpi'] = 200
plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
def foo(a, minv, maxv):
	if a > maxv:
		return maxv
	if a < minv:
		return minv
	return a
```

```python
from random import random

noise_lvl = 50
amount = 10

noised_images = [img] * amount
noised_images = np.array(noised_images)
for noised_image in noised_images:
	for x in range(img.shape[0]):
		for y in range(img.shape[1]):
			noise = np.random.random() * noise_lvl
			noised_image[x, y] = foo(noised_image[x, y] + noise, 0, 255)
# # display(Image.fromarray(noised_images[randint(0, amount)]))
plt.imshow(noised_images[5], cmap='gray')
plt.axis('off')
```

```python
clear_image = np.zeros_like(img)

clear_image = np.mean(noised_images, axis=0)
plt.rcParams['figure.dpi'] = 200
plt.imshow(clear_image, cmap='gray')
plt.axis('off')
```
