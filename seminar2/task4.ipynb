```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = np.array(cv.cvtColor(cv.imread('Lena.png'), cv.COLOR_BGR2RGB))
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
# hist = cv.calcHist([img], [0], None, [256], [0, 256])
# plt.plot(hist)
# plt.show()
```

#  

```python
a = float(input("Input a: "))
b = float(input("Input b: "))
corrected1 = img.copy()
for i in range(img.shape[0]):
	for j in range(img.shape[1]):
		corrected1[i, j] = np.clip(a * img[i, j] + b, 0, 255)

plt.imshow(corrected1, cmap='gray')
plt.axis('off')
plt.show()
# hist = cv.calcHist([corrected1], [0], None, [256], [0, 256])
# plt.plot(hist)
# plt.show()
```

#  

## $\gamma > 1$

```python
gamma = 7
corrected2 = img.copy()

for i in range(img.shape[0]):
	for j in range(img.shape[1]):
		corrected2[i, j] = ((img[i, j] / 255) ** gamma) * 255

plt.imshow(corrected2, cmap='gray')
plt.axis('off')
plt.show()
# hist = cv.calcHist([corrected2], [0], None, [256], [0, 256])
# plt.plot(hist)
# plt.show()
```

## $\gamma < 1$

```python
gamma = 0.5
corrected3 = img.copy()
for i in range(img.shape[0]):
	for j in range(img.shape[1]):
		corrected3[i, j] = ((img[i, j] / 255) ** gamma) * 255

plt.imshow(corrected3, cmap='gray')
plt.axis('off')
plt.show()
# hist = cv.calcHist([corrected3], [0], None, [256], [0, 256])
# plt.plot(hist)
# plt.show()
```

```python
img = np.array(cv.cvtColor(cv.imread('Lena.png'), cv.COLOR_BGR2GRAY))
equalized = img.copy()
hist = cv.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
equalized = cv.equalizeHist(equalized)
equalized_hist = cv.calcHist([equalized], [0], None, [256], [0, 256])
plt.plot(equalized_hist)
plt.show()
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
plt.imshow(equalized, cmap='gray')
plt.axis('off')
plt.show()
```
